Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

The ArithmeticError class in Python represents errors raised for numeric calculation errors. Two of its subclasses are ZeroDivisionError and OverflowError.

ZeroDivisionError:Raised when attempting to divide by zero.



ZeroDivisionError: Raised when attempting to divide by zero.

```python
try:
    result = 5 / 0
except ZeroDivisionError as e:
    print("Error:", e)
```

OverflowError: Raised when the result of an arithmetic operation is too large to be represented.

```python
import math
try:
    result = math.exp(1000)
except OverflowError as e:
    print("Error:", e)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.


The LookupError class in Python acts as a base class for exceptions raised when a key or index is not found. Two of its common subclasses are KeyError and IndexError.

KeyError: Raised when a dictionary key is not found.


```python

my_dict = {'a': 1, 'b': 2}

try:
    value = my_dict['c']
except KeyError as e:
    print("Error:", e)
```

 IndexError: Raised when a sequence index is out of range.

```python
my_list = [1, 2, 3]
try:
    value = my_list[5]
except IndexError as e:
    print("Error:", e)
```

Q6. List down some best practices for exception handling in python.

Some of the best practices for exception handling in python are 

Specificity: Use specific exceptions when handling errors.
Try-Except blocks: Wrap code that might raise an exception in a try-except block.

Avoid overly broad exceptions: Catch only the exceptions you expect and handle them appropriately.

Cleanup with finally: Use finally block to ensure resources are properly released.
Logging: Log exceptions to aid debugging and analysis.
Keep it readable: Make your exception handling clear and understandable.
Handle exceptions at the right level: Handle exceptions where it makes sense in your code hierarchy.
