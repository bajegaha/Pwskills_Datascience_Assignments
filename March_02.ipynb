<B>Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.</B>


Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in data analysis and presentation. It provides a wide range of tools for constructing plots, charts, histograms, and other graphical representations of data.<br>
The Pyplot module in Matplotlib provides a MATLAB-like interface for creating a variety of plots. Here are five types of plots that can be created using Pyplot:
<B>Line Plot</B>: Displays data points connected by straight line segments. It's great for showing trends over time or relationships between variables.<br>
<B>Scatter Plot</B>: Represents individual data points as markers. It's useful for visualizing the relationship between two variables to identify patterns or correlations.<br>
<B>Histogram</B>: Shows the distribution of a single variable by dividing data into bins and displaying the frequency of occurrences in each bin.<br>
<B>Bar Plot</B>: Displays categorical data with rectangular bars. It's effective for comparing values across different categories.<br>
<B>Pie Chart</B>: Represents the proportions of categorical data as slices of a circle. It's commonly used to illustrate percentages or proportions of a whole.

A scatter plot is a type of plot that displays values for two variables as a collection of points. Each point on the plot represents the value of both variables, making it useful for identifying relationships or patterns between two continuous variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

The subplot() function in Matplotlib is used to create multiple plots within the same figure. It's particularly helpful when you want to compare different datasets or display various aspects of data side by side.


```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


# Creating subplots
plt.figure(figsize=(10, 4))

# Subplot 1
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.plot(x1, y1)
plt.title('Line 1')

# Subplot 2
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot
plt.plot(x2, y2)
plt.title('Line 2')

# Subplot 3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot
plt.plot(x3, y3)
plt.title('Line 3')

# Subplot 4
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot
plt.plot(x4, y4)
plt.title('Line 4')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

<br><br><B>Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.</B>

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A bar plot is a graphical representation of categorical data with rectangular bars. The length or height of each bar is proportional to the value it represents. It's used to compare and visualize the quantitative values of different categories or groups.

A bar plot is useful for:<br>
<B>Comparison</B>: Comparing values between different categories.<br>
<B>Trend Analysis</B>: Displaying changes in data over time or categories.<br>
<B>Ranking</B>: Showing the rank or order of different categories based on a specific metric.


```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar plot
plt.figure(figsize=(8, 3))
plt.bar(company, profit, color='green')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company Profits (Bar Plot)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 3))
plt.barh(company, profit, color='blue')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.tight_layout()
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a summary of key statistical measures, such as the median, quartiles, and potential outliers. The plot consists of a rectangular "box" that represents the interquartile range (IQR) and "whiskers" that extend from the box to show the range of the data. Outliers may be plotted individually.

Box plots are used for:

<B>Identifying Central Tendency</B>: Displaying the median and quartiles to understand the central tendency of the data.<br>
<B>Detecting Outliers</B>: Identifying potential outliers or extreme values in the dataset.<br>
<B>Comparing Distributions</B>: Comparing the distribution of multiple datasets.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Values')
plt.show()
```
